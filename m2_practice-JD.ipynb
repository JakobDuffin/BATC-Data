```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('companies.csv')

print(df.info())
df.head()
```

```python
# Question 1 & 2

sns.countplot(x='state', data=df, width=.8)

plt.title('Company Count By State')
plt.xticks(size=7, rotation=60)

plt.show()
```

```python
# Question 3 & 4

openings_by = df.groupby(['Industry'], as_index=False).agg('job_openings').sum()
openings_by.sort_values(by='job_openings', ascending=False, inplace=True)

sns.barplot(data=openings_by, x='job_openings', y='Industry')

plt.title('Total Job Openings By Company')

plt.show()
```

```python
# Question 5 & 6

growth_found = df.groupby(['founded'], as_index=False).agg('growth_percentage').mean()

sns.lineplot(data=growth_found, x='founded', y='growth_percentage')

plt.title('Average Growth By Year Founded')
plt.xlabel('Year Founded')
plt.ylabel('Average Growth')

#growth_found.sort_values(by='founded')

plt.show()
```

```python
# Question 7 & 8

sns.scatterplot(data=df, x='employees', y='estimated_revenues')

plt.title('Employee Count & Estimated Revenue Comparison')
plt.xlabel('Number Of Employees')
plt.ylabel('Estimated Revenue (x1,000,000,000)')

plt.show()
```

```python
# Question 9 & 10

sns.histplot(data=df, x='founded')

plt.title('Number Of Companies Founded By Year')
plt.xlabel('Year Founded')

plt.show()
```
